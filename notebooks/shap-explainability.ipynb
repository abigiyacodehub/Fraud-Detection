## Stakeholder interpretation

- Use global explanations to identify the dominant fraud drivers and remove noisy or redundant features.
- Use local explanations to give analysts transaction-level reasons for high-risk flags.
- Review sensitive or proxy features for bias and fairness risks before deployment.
- Treat SHAP as model behavior explanation, not proof of causal relationships.

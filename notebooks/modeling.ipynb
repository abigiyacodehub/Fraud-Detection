# Interim 2 Modeling

Task 2 evidence notebook for stratified train-test splits, Logistic Regression baseline, tuned Random Forest ensemble, model comparison, and saved model artifacts.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(ROOT))

from scripts.train_task2_models import prepare_ecommerce, prepare_creditcard, train_models_for_dataset
from src.modeling import results_to_frame
```

## Data preparation

Both datasets are prepared with the Task 1 cleaning and feature engineering pipeline. The training function applies a stratified train-test split for each dataset.

```python
ecommerce_X, ecommerce_y, ecommerce_num, ecommerce_cat = prepare_ecommerce()
credit_X, credit_y, credit_num, credit_cat = prepare_creditcard()

ecommerce_y.value_counts(normalize=True), credit_y.value_counts(normalize=True)
```

## Baseline and ensemble training

The baseline model is Logistic Regression. The ensemble model is Random Forest with `GridSearchCV` hyperparameter tuning. Evaluation includes AUC-PR, F1-score, and confusion matrix values.

```python
all_results = []
split_reports = []

for dataset_name, X, y, numeric_features, categorical_features in [
    ('Ecommerce', ecommerce_X, ecommerce_y, ecommerce_num, ecommerce_cat),
    ('CreditCard', credit_X, credit_y, credit_num, credit_cat),
]:
    results, split_report = train_models_for_dataset(dataset_name, X, y, numeric_features, categorical_features)
    all_results.extend(results)
    split_reports.append(split_report)

comparison = results_to_frame(all_results)
comparison
```
